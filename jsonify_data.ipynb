```python
import pandas as pd
```

```python
col_data= pd.read_csv("plotting.csv")
```

```python
col_data.shape
```

```python
col_data.head()
```

```python
import json
json_data=[]
temp_data={}
data=[]
dict_json={}
for index,row in col_data.iterrows():
    location_info={
        "lat":row["lat"],
        "lng":row["lon"],
        "state":row["State"]
    }
    salary_info={ 
        "mid-career-median-salary":row["Alumni Salary"],
        "rank":row["Rank"],
        "code":row["Code"]

    }
    score_info={
        "SAT Lower":row["SAT Lower"], 
        "SAT Upper":row["SAT Upper"], 
        
    }
    admissions_info={
        "students":row["Student Population"],
        "Acceptance Rate":row["Acceptance Rate"], 
    }
    json_data.append({
        "location_info":location_info,
        "school":row["Name"],
        "link":row["Website"],
        "salary_info":salary_info,
        "score_info":score_info,
        "admissions_info":admissions_info
        
    })
    
```

```python
json_data
```

```python
with open("UNIVERSITIES.json", 'w') as outfile:
    outfile.write(json.dumps(json_data))
```
